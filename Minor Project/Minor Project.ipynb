```python
data_csv_maharashtra = pd.read_csv('covid_19_india.csv', index_col = 1, parse_dates = True, dayfirst = True)
start_date = '2020-06-01'
end_date   = '2020-06-30'
data_csv_maharashtra = data_csv_maharashtra.loc[start_date : end_date]
data_csv_maharashtra2 = data_csv_maharashtra.loc[(data_csv_maharashtra['State/UnionTerritory'] == 'Maharashtra')]
data_csv_maharashtra = pd.DataFrame(data_csv_maharashtra2)
data_csv_maharashtra = data_csv_maharashtra.reset_index()
data_csv_maharashtra.get(['Sno', 'Date', 'Time', 'State/UnionTerritory', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed'])
for data in data_csv_maharashtra.index:
    data_csv_maharashtra['Sno'][data], data_csv_maharashtra['Date'][data], data_csv_maharashtra['Time'][data], data_csv_maharashtra['State/UnionTerritory'][data], data_csv_maharashtra['ConfirmedIndianNational'][data], data_csv_maharashtra['ConfirmedForeignNational'][data], data_csv_maharashtra['Cured'][data], data_csv_maharashtra['Deaths'][data], data_csv_maharashtra['Confirmed'][data]
data_csv_maharashtra
```

# Data Visualization Using Matplotlib

```python
dateList = [data.strftime("%d") for data in data_csv_maharashtra['Date'].tolist()]
pl.plot(dateList, data_csv_maharashtra['Confirmed'].tolist(), color = 'green', label = 'Confirmed', linewidth = 5)
pl.plot(dateList, data_csv_maharashtra['Cured'].tolist(), color = 'blue', label = 'Cured', linewidth = 5)
pl.plot(dateList, data_csv_maharashtra['Deaths'].tolist(), color = 'red', label = 'Deaths', linewidth = 5)
pl.title('COVID - 19 Maharashtra')
pl.xlabel('Days Of June')
pl.ylabel('Number Of People')
pl.legend()
pl.grid(True, color = 'gray')
pl.legend(loc = 'best')
pl.show()
```

# Neighbouring States

```python
state_list = np.array(state_csv['State'])
state_list
```

```python
confirmed = state_csv['Confirmed'].to_numpy()
confirmed
```

```python
confirmed_case = dict(zip(state_list, confirmed))
confirmed_case
```

```python
recovered = state_csv['Recovered'].to_numpy()
recovered
```

```python
recovered_case = dict(zip(state_list, recovered))
recovered_case
```

```python
deaths = state_csv['Deaths'].to_numpy()
deaths
```

```python
deaths_case = dict(zip(state_list, deaths))
deaths_case
```

```python
active = state_csv['Active'].to_numpy()
active
```

```python
active_case = dict(zip(state_list, active))
active_case
```

```python
neighbouring_state_list = np.array(neighbouring_state_csv['Neighbouring State'])
neighbouring_state_list
```

```python
neighbouring = dict(zip(state_list, neighbouring_state_list))
neighbouring
```

```python
enter_state = input("Enter the State:").title()
for state in state_list:
    if state.strip() == enter_state.strip():
        neighbour = neighbouring[state].split(',')
        print('Neighbouring States:', neighbour)
        neighbours = np.arange(len(neighbour))
        confirm_case = []
        recover_case = []
        death_case = []
        actived_case = []
        for i in neighbour:
            c_case = confirmed_case[i]
            r_case = recovered_case[i]
            d_case = deaths_case[i]
            a_case = active_case[i]
            confirm_case.append(c_case)
            recover_case.append(r_case)
            death_case.append(d_case)
            actived_case.append(a_case)
        print('Confirmed Case in Neighbouring States:', confirm_case)
        print('Recovered Case in Neighbouring States:', recover_case)
        print('Deaths Case in Neighbouring States:', death_case)
        print('Active Case in Neighbouring States:', actived_case)
        style.use('ggplot')
        pl.figure(figsize = (20,10))
        pl.xticks(neighbours, neighbour, fontsize = 15)
        pl.yticks(fontsize = 15)
        width = 0.2
        pl.bar(neighbours, confirm_case, color = 'green', label = 'Confirmed', width = 0.2)
        pl.bar(neighbours + width, recover_case, color = 'blue', label = 'Recoverd', width = 0.2)
        pl.bar(neighbours + width + width, death_case, color = 'red', label = 'Deaths', width = 0.2)
        pl.bar(neighbours + width + width + width, actived_case, color = 'gray', label = 'Active', width = 0.2)
        pl.title('COVID - 19 Report of ' + state + ' Neighbouring State', fontsize = 25)
        pl.xlabel('Days Of June', fontsize = 20)
        pl.ylabel('Number Of People', fontsize = 15)
        pl.legend(fontsize = 20)
        pl.show()
        break
else:
    print('SORRY:)!!,You Enter Wrong State...')
```
